```python
# Importando a biblioteca Pandas para trabalhar com os dados.
import pandas as pd

# Armazenando o caminho do arquivo em uma variável.
dados_caminho = '/content/dados.csv'

# Lendo o conteúdo de um arquivo CSV e atribuindo-o a uma variável.
dados = pd.read_csv(dados_caminho, sep=';', engine='python', encoding='utf-8')
```

```python
# Verificando os dados importados.

# Imprimindo informações gerais sobre o conjunto de dados.
display(dados.info())

# Imprimindo as primeiras 10 linhas do arquivo.
display(dados.head(10))
# Imprimindo as últimas 10 linhas do arquivo.
display(dados.tail(10))
```

```python
# Criando uma nova variável com a cópia do conjunto de dados original.
  # Utilizou-se o método "copy()", que cria uma cópia exata do DataFrame original.
  # Caso fosse simplesmente utilizada a atribuição "dados_copia = dados" a nova variável seria apenas uma referência ao DataFrame original.
    # Como consequência qualquer modificação no conteúdo de uma variável afetaria o conteúdo da outra.
dados_copia = dados.copy()
```

```python
# Na nova variável:

# Substituindo os valores nulos (NaN) da coluna "Calories" por 0 (zero).
dados_copia['Calories'].fillna(0, inplace=True)
# Imprimindo o conjunto de dados para verificar a mudança.
display(dados_copia)
```

```python
# Ainda na nova variável:

# Substituindo os valores nulos (NaN) da coluna "Date" por "1900/01/01".
dados_copia['Date'].fillna('1900/01/01', inplace=True)
# Imprimindo o conjunto de dados para verificar a mudança.
display(dados_copia)
# Transformando os dados da coluna "Date" em datetime usando o método "to_datetime()".
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'])

# Isso deverá gerar um erro: o valor "1900/01/01" não corresponde ao formato "'%Y/%m/%d'".
```

```python
# Corrigindo o erro anterior:

# Utilizaremos a biblioteca "numpy" para utilizar o valor "Not a Number" (NaN).
import numpy as np

# Substituindo, na coluna "Date", o valor "1900/01/01" por "NaN".
dados_copia.loc[dados_copia['Date'] == '1900/01/01', 'Date'] = np.nan
# Utilizando o método "to_datetime" para tranformar a coluna "Date" em "datetime".
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'])
# Imprima o conjunto de dados para verificar se as mudanças acima foram aplicadas com sucesso.
display(dados_copia)

# Isso deverá gerar outro erro: o valor "20201226" não corresponde ao formato "'%Y/%m/%d'".
```

```python
# Corrigindo o novo erro:

# Transformando o valor "20201226" para a string "2020-12-26" usando o método replace.
  # Garante que o valor a ser substituído seja tratado como string para evitar problemas de tipo.
dados_copia['Date'] = dados_copia['Date'].replace('20201226', '2020/12/26')

# Utilizando o método "to_datetime" para transformar a coluna "Date" em "datetime".
  # Adicionando 'format="mixed"' para lidar com diferentes formatos de data na coluna e 'errors="coerce"' para transformar valores inválidos em NaT (Not a Time).
dados_copia['Date'] = pd.to_datetime(dados_copia['Date'], format='mixed', errors='coerce')

# Exibindo o DataFrame para verificar as alterações
display(dados_copia)
```

```python
# Removendo os registros que contêm valores nulos na coluna 'Date'.
  # O método 'dropna(subset=['Date'])' remove as linhas onde a coluna 'Date' possui valores nulos (NaT).
dados_copia = dados_copia.dropna(subset=['Date'])

# Exibindo o DataFrame resultante para verificar a remoção da linha.
display(dados_copia)
```
